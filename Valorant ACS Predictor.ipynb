```python
#importation 
import pandas as pd 
import numpy as np 
from sklearn.preprocessing import StandardScaler

print("--- Bloc1 : Chargement et Inspection ---")

# 1. Génération de notre base de données
np.random.seed(42)
n_parties=500


# On génère des statistiques réalistes pour 500 parties
adr = np.random.normal(140,30,n_parties) #Degats : moyenne 140
hs_percent=np.random.normal(20,5,n_parties) #Headshots : moyenne 20%
flashes=np.random.poisson(8,n_parties)

# On crée la cible (ACS) via une formule mathématique cachée avec du "bruit" aléatoire

acs=(adr*1.5) +(hs_percent*2.5)+(flashes*1.2) +np.random.normal(0,15,n_parties)


# 2. Création du DataFrame Pandas (Notre table de données)
df = pd.DataFrame({
    "ADR":adr , 
    "HS_Percent":hs_percent , 
    "Flashes_Reussies" : flashes , 
    "Acs_Cible":acs
})


# 3. L'Exploration
print("\n1. Apercu des 5 premiers lignes ") 
print(df.head())

print("\n2. Informations sur les colonnes")
print(df.info())


print("\n3. Statistiques descriptives (Moyennes, min, max) :")
print(df.describe().round(1))
```

```python
#BLOC 2 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression 
X=df[['ADR','HS_Percent','Flashes_Reussies']]
Y=df['Acs_Cible']
#Appique de standard Scaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)

X_scaled
```

```python
#Bloc 3 
X_train , X_test , y_train,y_test =train_test_split(X_scaled , Y ,test_size=0.2 , random_state=42 ) 

print(f"taille de donner dentrainement :{X_train.shape[0]} lignes ")
print(f"Taille des données d'examen (Test)        : {X_test.shape[0]} lignes")
```

```python
#Bloc 4 Modelisation et entrainement 

print("\n-- Bloc 4 :Entrainement de lalgorithme ---")
model_regression=LinearRegression ()

model_regression.fit(X_train,y_train)

print("entrainement reussie")
```

```python
# 3. Interprétation immediate : les poids découverts par la machine 
colonnes =X.columns 
poids_decouverts=model_regression.coef_
for col, poids in zip(colonnes, poids_decouverts):
    print(f"L'impact de la variable {col} sur l'ACS est de : {poids:.2f}")
print(f"Le Biais (Ordonnée à l'origine) est de : {model_regression.intercept_:.2f}")
```

```python
#Bloc5 
from sklearn.metrics import mean_squared_error, r2_score

print("\n--- BLOC 5 : evaluation et performance ---")

y_pred = model_regression.predict(X_test)

# 2. Calcul des metriques d'evaluation
rmse_final = np.sqrt(mean_squared_error(y_test, y_pred))
r2_final = r2_score(y_test, y_pred)

print(f"Score R2 Final : {r2_final:.3f}")
print(f"RMSE Final     : {rmse_final:.2f} points d'ACS")
```
